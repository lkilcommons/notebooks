# Throwing Up and Down

A person standing at the edge of a cliff throws one ball straight up and another ball straight down at the same initial speed. Neglecting air resistance, the ball to hit the ground below the cliff with the greater speed is the one initially thrown

* A) upward.
* B) downward.
* C) neither- they both hit at the same speed.

## Solution
This is, at it's heart, an acceleration problem. Specifically, it's a **constant** acceleration problem, because gravity always subjects an object that is not being acted on by other forces to the same acceleration: 

$$g = 9.81 \frac{m}{s^2}$$

What do we know?

1. Both balls *start* at the same position, and *end* at the same position. That is, they go through the same *displacement*. 
2. Both balls have the same intial **speed** but are thrown in opposite directions

The important thing to realize is that, even though the ball that travels upward moves for longer, it starts and ends at the same position as the ball that moves downward, and the difference between initial position and final position is all that ideal kinematics cares about. It doesn't give a crap what happens in between intital and final positions because *the forces on the balls don't change during that time*

**If we didn't ignore air resistance this would be a different story**
## The mathmatical representation of 1 is:

* At the top of the cliff both are at ($y = H$) if the cliff is $H$ meters high
$$ y_{1,i} = y_{2,i} = H $$

* At the bottom of the cliff they are both on the ground
$$ y_{1,f} = y_{2,f} = 0 $$

* So,
$$ y_{1,f}-y_{1,i} = H - 0 = H $$

$$ y_{2,f}-y_{2,i} = H - 0 = H $$

* So,

$$ \Delta y_{1} = \Delta y_{2} = H$$

## 2 is a bit more simple to state mathmatically
* Let's say both balls start at speed $S$ but in opposite directions:

$$ v_{1,i} = S $$
$$ v_{2,i} = -S $$

* So,

$$ v_{1,i} = - v_{2,i} $$

## Kinematics equation

* We need an equation that tells us about changes in position

$$ v_f^{2} = v_i^{2} + 2g(y_f - y_i) $$ 

$$ v_f^{2} = v_i^{2} + 2g\Delta y $$


* Well how convenient! 

* $\bf{v_i}$ is squared meaning it **doesn't matter if it's positive or negative**

## Proving it

* These are the equations of motion for both balls

$ v_{1,f}^{2} = v_{1,i}^2 + 2g\Delta y_{1} $

$ v_{2,f}^{2} = v_{2,i}^2 + 2g\Delta y_{2} $

* Now we plug in what we know from 1

$ v_{1,f}^{2} = v_{1,i}^2 + 2gH $

$ v_{2,f}^{2} = v_{2,i}^2 + 2gH $

* Now we plug in 2

$ v_{1,f}^{2} = (S)^2 + 2gH $

$ v_{2,f}^{2} = (-S)^2 + 2gH $ 

* So,


$ v_{1,f}^{2} = S^2 + 2gH $

$ v_{2,f}^{2} = S^2 + 2gH $

* Take square roots,

$ v_{1,f} = \sqrt{S^2 + 2gH} $

$ v_{2,f} = \sqrt{S^2 + 2gH} $




```python
#We can prove it using the other kinematics equations in a few lines of code

import numpy as np
from matplotlib import pyplot as pp

def calc_yf(yi,vi,t): 
    yf = yi + vi*t + 1./2*(-9.81)*(t**2.)
    #find out when hits ground
    t_g = t[np.nanargmin(np.abs(yf))]
    return yf,t_g

def calc_vf(yi,vi,t):
    vf = vi-(9.81)*t
    return vf

H = 20.
S = 8.

y1i = H
y2i = H
v1i = S
v2i = -1*S

ti = 0
tf = 3

t = np.linspace(ti,tf,100)

y1f,tg1 = calc_yf(y1i,v1i,t)
y2f,tg2 = calc_yf(y2i,v2i,t)

v1f = calc_vf(y1i,v1i,t)
v2f = calc_vf(y2i,v2i,t)


f = pp.figure(figsize=(12,4),dpi=300)
a1 = f.add_subplot(121)
a2 = f.add_subplot(122)
a1.set_title('Height')
a1.plot(t,y1f,'b.',label='Ball 1')
a1.plot(t,y2f,'r.',label='Ball 2')
a1.axhline(H,label='Cliff',color='purple')
a1.axhline(0.,label='Ground',color='black')
a1.legend()
a1.grid(True)
a1.set_xlabel('Time (seconds)')
a1.set_ylabel('Height (meters)')

a2.set_title('Speed')
a2.plot(t,v1f,'b.',label='Ball 1')
a2.plot(t,v2f,'r.',label='Ball 2')
a2.axvline(tg1,label='Ball 1 hits', color='blue')
a2.axvline(tg2,label='Ball 2 hits', color='red')
a2.set_xlabel('Time (seconds)')
a2.set_ylabel('Speed (meters per second)')


a2.legend()
a2.grid(True)

pp.show()
```
